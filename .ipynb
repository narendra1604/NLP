```python
doc = nlp("DR. starnge loves pav bhaji of mumbai as it cost only 2$ per plate.")
for token in doc:
    print(token, " | ", token.pos_, " | ", token.lemma_,"|")
```

```python
doc = nlp("Tesla Inc is going to acquire for $45 billion")

for ent in doc.ents:
    print(ent.text,"|", ent.label_)
```
